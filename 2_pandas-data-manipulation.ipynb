```python
import pandas as pd

df = pd.read_csv('data\\train.csv')
df.head(2)
```

## Indexing by position

```python
# Indexing by a specific column, say here we want the name
col = df.iloc[:,3]
col.head(5)
```

```python
# this could also be written as
df.iloc[:5, 3]
```

```python
# this can also be done using the name of the column
df.loc[:4, "Name"]
```

```python
# say we want to display the index 0 to index 6 from name
df.loc[:6, "Name"]
```

```python
# print the first 7 records from name to age
df.loc[:6, "Name":"Age"]
```

```python
# Let's add another column to see functions
df['Actions'] = 1
```

```python
df.head(2)
```

```python
# Now, to set value to the entire column
df['Actions'] = 5
df.loc[:2, "Actions"]
```

```python
# Now adding function to the column
func = lambda x: x * 2
df["Actions"] = df["Actions"].apply(func)
df.head(2)
```

```python
# Sort by column value in descending order
df.sort_values(by="Age", ascending = False)
```

```python
# Filtering the data
filters = df['Age'] > 70
df_new = df[filters]
df_new
```

```python
# now say we want filter such that the age > 70 and Pclass < 2
filters1 = (df['Age'] > 70) & (df['Pclass'] < 2)
df_new1 = df[filters1]
df_new1
```
